```python
df2 = df1[df1['vote_count'].notnull() & df1['vote_average'].notnull()]
```

```python
C = df2['vote_average'].mean()
m = df2['vote_count'].quantile(0.95)
```

```python
top_movies = df2[df2['vote_count'] >= m]

top_movies.loc[:, 'weighted_rating'] = (top_movies['vote_count'] / (top_movies['vote_count'] + m)) * top_movies['vote_average'] + (m / (top_movies['vote_count'] + m)) * C


top_movies = top_movies.sort_values('weighted_rating', ascending=False)[['title','year', 'vote_count', 'vote_average', 'popularity','weighted_rating']]

overall_top_250 = top_movies.head(250)
overall_top_250.head()
```

```python
def genre_top_chart(genre):
    genre_movies = df2[df2['genres'].apply(lambda x: genre in x)]
    genre_top_movies = genre_movies[genre_movies['vote_count'] >= m]
    genre_top_movies.loc[:,'weighted_rating'] = (genre_top_movies['vote_count'] / (genre_top_movies['vote_count'] + m)) * genre_top_movies['vote_average'] + (m / (genre_top_movies['vote_count'] + m)) * C
    genre_top_movies = genre_top_movies.sort_values('weighted_rating', ascending=False)
    return genre_top_movies[['title','year', 'vote_count', 'vote_average', 'popularity','weighted_rating']]
```

```python
# Extract director's name from credits.csv
def get_director_name(crew):
    crew_list = ast.literal_eval(crew)
    for member in crew_list:
        if member['job'] == 'Director':
            return member['name']
    return ''

# Add director column to movies_df
df_small['director'] = df_small['crew'].apply(get_director_name)
```

```python
def get_top_actors(cast):
    cast_list = ast.literal_eval(cast)
    actors = []
    for i in range(min(5, len(cast_list))):
        actors.append(cast_list[i]['name'])
    return actors

# Add actors column to movies_df
df_small['actors'] = df_small['cast'].apply(get_top_actors)
```

```python
df_small['keywords'] = df_small['keywords'].apply(ast.literal_eval)

# Flatten the list of dictionaries and stem the keywords
stemmer = PorterStemmer()
keywords_list = [stemmer.stem(keyword['name']) for keywords in df_small['keywords'] for keyword in keywords]

# Count the frequency of each keyword
keyword_frequency = Counter(keywords_list)

# Filter keywords with frequency greater than 1
filtered_keywords = {keyword: freq for keyword, freq in keyword_frequency.items() if freq > 1}
```

```python
def filter_keywords(keywords):
    return [stemmer.stem(keyword['name'].replace(" ", "").lower()) for keyword in keywords if stemmer.stem(keyword['name'].replace(" ", "").lower()) in filtered_keywords]

# Apply the filter_keywords function to keep only the filtered keywords in the DataFrame
df_small['keywords'] = df_small['keywords'].apply(filter_keywords)
```

```python
def get_recommendations(title, cosine_sim=linear_sim, df=df_content):
    # Get the index of the movie title
    indices = pd.Series(df.index, index=df['title']).drop_duplicates()
    idx = indices[title]

    # Compute the pairwise similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top 10 similar movies
    top_movies_indices = [i[0] for i in sim_scores[1:11]]
    top_movies = df['title'].iloc[top_movies_indices]
    return top_movies
```

```python
df_popularity = df_content
df_popularity['vote_count'] = df_small['vote_count']
df_popularity['vote_average'] = df_small['vote_average']
```

```python
def get_popular_recommendations(title, linear_sim=linear_sim, df=df_popularity):
    # Get the index of the movie title
    indices = pd.Series(df.index, index=df['title']).drop_duplicates()
    idx = indices[title]

    sim_scores = list(enumerate(linear_sim[idx]))

    # Sort the movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top 30 similar movies
    top_movies_indices = [i[0] for i in sim_scores[1:31]]
    top_movies = df[['title','popularity_score']].iloc[top_movies_indices]
    return top_movies.sort_values('popularity_score',ascending = False).head(10)['title']
```

```python
vote_counts = df_popularity['vote_count']
vote_averages = df_popularity['vote_average']
C = vote_averages.mean()
m = vote_counts.quantile(0.90)
df_popularity['popularity_score'] = (vote_counts / (vote_counts + m)) * (vote_averages - C)
```

```python
recommendations = get_popular_recommendations(movie_title)
recommendations
```

```python
# Load the ratings data into Surprise Dataset
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Build the full training set
trainset = data.build_full_trainset()

# Choose a collaborative filtering algorithm (e.g., SVD)
algo = SVD()

# Train the model on the entire dataset
algo.fit(trainset)

# Movie ID for which you want to find the average rating
movie_id = 31

# Predict ratings for the specified movie
movie_predictions = algo.predict(uid=None, iid=str(movie_id))

# Average rating of the movie based on the predicted ratings
average_rating = round(movie_predictions.est, 2)
print(f"Average rating for Movie ID {movie_id}: {average_rating}")
```

```python
# Merge the DataFrames based on the 'movieId' column
df_merged = ratings.merge(link_map, on='movieId', how='left')

df_merged.head()
```

```python
df_merged.shape
```

```python
df_merged['tmdbId'] = df_merged['tmdbId'].astype('Int64')
```

```python
df1['id'].head()
```

```python
def get_collab_recommendations(user_id):
    # Get the list of movie IDs
    movie_ids = ratings['movieId'].unique()

    # Predict ratings for all movies for the specified user
    predictions = [algo.predict(user_id, str(movie_id)) for movie_id in movie_ids]

    # Sort the predictions based on estimated ratings
    sorted_predictions = sorted(predictions, key=lambda x: x.est, reverse=True)

    # Get the top N movie recommendations
    top_movies = [pred.iid for pred in sorted_predictions[:10]]
    df = df_merged[df_merged['movieId'].isin(top_movies)]
    id_list = list(df_merged['tmdbId'])
    top_movies = list(df1[df1['id'].isin(id_list)]['title'])
    return top_movies
```

```python
def get_hybrid_recommendations(user_id, movie_title):
    # Get collaborative filtering recommendations for the user
    collab_rec = get_collab_recommendations(user_id)
    # Get content-based filtering recommendations for the movie
    content_rec = get_popular_recommendations(movie_title)

    # Merge the two recommendation lists and remove duplicates
    hybrid_rec = list(set(collab_rec + list(content_rec)))

    return hybrid_rec[:10]
```
